```python
from numpy import *
from matplotlib.pyplot import *
```

```python
N = 50
xrange = linspace(-2.,2.,N)
fx = lambda x: x*x
def sx(x):
    ss = 4./3.
    for n in range(1, 10):
        ss = ss + 16 * (-1)**n * np.cos( n*np.pi*x/2) / (n*n*np.pi*np.pi)
    return ss

plot(xrange,fx(xrange),color='blue',label='f(x)')
plot(xrange,sx(xrange),color='green',label='S(x)')
grid(True)
legend(loc='lower right')

```
